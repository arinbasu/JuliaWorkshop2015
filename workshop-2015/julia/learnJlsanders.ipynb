```julia
data = DataFrame()
```

```julia
gender_array = ["fem"]
```

```julia
for n = 1:199
    rn = randn()
    if (rn <= 0)
    append!(gender_array, ["fem"])
    else 
        append!(gender_array, ["male"])
    end
end
```

```julia
gender_array
```

```julia
data[:gender] = gender_array
```

```julia
group_array = ["A"]
for n = 1:199
    rn = randn()
    if (rn <= 0)
        append!(group_array, ["B"])
    else
        append!(group_array, ["A"])
    end
end
```

```julia
group_array
```

```julia
data[:Group] = group_array
```

```julia
Pkg.add("Distributions")
using Distributions
data[:Age] = rand(Normal(35, 10), 200)
data[:Days] = rand(Poisson(2), 200)
data[:Temp] = rand(Normal(38.5, 2), 200)
data[:WCC] = rand(Normal(15, 5), 200)
data[:CRP] = rand(Poisson(2), 200) * 2 * 100
```

```julia
data
```

```julia
Gadfly.plot(data, x ="WCC", y = "CRP",
Geom.point, Guide.title("Kernel Density Plot for wcc"))
```

```julia
## sub dataframes
groupA = data[data[:Group] .== "A", :]
groupB = data[data[:Group] .== "B", :]
```

```julia
mean(groupA[:WCC])
```

```julia
std(groupA[:WCC])
```

```julia
onesamp = EqualVarianceTTest(groupA[:WCC], groupB[:WCC])
```

```julia
methods(onesamp)
```

```julia
pvalue(onesamp)
```

```julia
ci(onesamp)
```
